```python
import pandas as pd
import cufflinks as cf
from IPython.display import display,HTML

# Tratamiento de datos
# ==============================================================================
import pandas as pd
import numpy as np

# Gráficos
# ==============================================================================
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
cf.set_config_file(sharing='public',theme='white',offline=True) # write cf.getThemes() to find all themes available
```

```python
RD_PBI=pd.read_excel(r'C:\Users\G&T Asesores\Desktop\Indicadores\SUPNT1 TOTAL\PROYECTO MAYO\RANGO DEUDA PBI.xlsx')
RD_PBI
```

```python
RD_PBI = RD_PBI.pivot(index='PBI', columns='RANGO_DEUDA',
                                    values='EFECTIVIDAD')
RD_PBI
```

```python
RD_PBI.iplot(kind='line', xTitle='PBI (%)', yTitle='EFECTIVIDAD (%)',
                    title='EFECTIVIDAD POR RANGO DE DEUDA Y PBI')
```

```python
RD_PBI.plot(kind="line",fontsize=11,figsize=(20,6))
plt.xlabel("PBI (%)",fontsize=12)
plt.ylabel("EFECTIVIDAD (%)",fontsize=12)
plt.title("EFECTIVIDAD POR RANGO DE DEUDA Y PBI",fontsize=15)
plt.show()
```
